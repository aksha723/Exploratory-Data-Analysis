```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Survived'].value_counts()
```

```python
# 1. Survived distribution (Bar plot)
sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = Dead, 1 = Survived)')
plt.show()




```

```python
# 2. Survived by Sex (Gender based survival)
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
 #Survived by Pclass (Ticket class based survival)
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Pclass')
plt.show()
```

```python
# 4. Survived distribution as Pie chart
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6), colors=['red', 'green'])
plt.title('Survival Proportion')
plt.ylabel('')
plt.show()
```

```python
# Sirf numeric columns select karo
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Correlation matrix calculate karo
corr_matrix = numeric_df.corr()

# Heatmap plot karo
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```
